### Simulaciones Simbólicas

```python
from pytc2.general import print_latex, print_subtitle, a_equal_b_latex_s
import sympy as sp
import pytc2.remociones as rmv
import pytc2.dibujar as dib
 
# Resolución simbólica
s = sp.symbols('s ', complex=True)

# función de excitación
ZZ=(s**4+7*s**2+10)/(3*s**3+7*s)

# remociones:    
print_subtitle("Remuevo C serie")
Z2, C1 = rmv.remover_polo_dc(ZZ, omega_zero = 1)
print_latex(a_equal_b_latex_s("Z_2",Z2))
print_latex(a_equal_b_latex_s("C_1",C1))

print_subtitle("Remuevo LC derivacion")
Y4, Y3, L2, C2 = rmv.remover_polo_jw(1/Z2, 1, isImpedance=False)
print_latex(a_equal_b_latex_s("Y_4",Y4))
print_latex(a_equal_b_latex_s("L_2",L2))
print_latex(a_equal_b_latex_s("C_2",C2))

print_subtitle("Remuevo C serie")
Z6, C3 = rmv.remover_polo_dc(1/Y4)
print_latex(a_equal_b_latex_s("Z_6",Z6))
print_latex(a_equal_b_latex_s("C_3",C3))

print_subtitle("Remuevo L serie")
Z8, L3 = rmv.remover_polo_infinito(Z6)
print_latex(a_equal_b_latex_s("Z_8",Z8))
print_latex(a_equal_b_latex_s("L_3",L3))

# dibujo red:
d = dib.Drawing(unit=4)
d = dib.dibujar_puerto_entrada(d,
                        voltage_lbl = ('+', '$V_1$', '-'), 
                        current_lbl = '$I_1$')
d, zz_lbl = dib.dibujar_funcion_exc_abajo(d, 
                                          '1/Y_{11}',  
                                          ZZ, 
                                          hacia_salida = True,
                                          k_gap_width = 0.5)
d = dib.dibujar_elemento_serie(d, dib.Capacitor, C1)
d = dib.dibujar_tanque_derivacion(d, L2, C2)
d = dib.dibujar_elemento_serie(d, dib.Capacitor, C3)
d = dib.dibujar_elemento_serie(d, dib.Inductor, L3)
d = dib.dibujar_puerto_salida(d,
                        voltage_lbl = ('+', '$V_2$', '-'), 
                        current_lbl = '$I2$')
dib.display(d)
```

#### Simulaciones Simbólicas

```python
from pytc2.general import print_latex, print_subtitle, a_equal_b_latex_s
import sympy as sp
import pytc2.remociones as rmv
import pytc2.dibujar as dib
 
# Resolución simbólica
s = sp.symbols('s ', complex=True)

# función de excitación
Y22=(s**2+6*s+8)/(s+3)

# remociones (desde atras hacia adelante xq son parametros Y):    
print_subtitle("Remuevo C paralelo")
Y2, C1 = rmv.remover_polo_infinito(Y22)
print_latex(a_equal_b_latex_s("Y_2",Y2))
print_latex(a_equal_b_latex_s("C_1",C1))

print_subtitle("Remuevo R derivacion")
Y4, G2 = rmv.remover_valor(Y2, 1)
R2 = 1/G2
print_latex(a_equal_b_latex_s("Y_4",Y4))
print_latex(a_equal_b_latex_s("R_2",R2))

print_subtitle("Remuevo RC serie")
Z6, ZRC, R3, C3 = rmv.remover_polo_sigma(1/Y4, 1, isImpedance=True)
print_latex(a_equal_b_latex_s("R_3",R3))
print_latex(a_equal_b_latex_s("C_3",C3))

print_subtitle("Remuevo R serie")
R4 = Z6
print_latex(a_equal_b_latex_s("R_4",R4))

# dibujo red:
d = dib.Drawing(unit=4)
d = dib.dibujar_puerto_entrada(d,
                        voltage_lbl = ('+', '$V_1$', '-'), 
                        current_lbl = '$I_1$')
d = dib.dibujar_elemento_serie(d, dib.Resistor, R4)
d = dib.dibujar_tanque_RC_serie(d, R3, C3)
d = dib.dibujar_elemento_derivacion(d, dib.Resistor, R2)
d = dib.dibujar_espacio_derivacion(d)
d = dib.dibujar_elemento_derivacion(d, dib.Capacitor, C1)
d = dib.dibujar_puerto_salida(d,
                        voltage_lbl = ('+', '$V_2$', '-'), 
                        current_lbl = '$I2$')

d, zz_lbl = dib.dibujar_funcion_exc_abajo(d, 
                                          'Y_{22}',  
                                          Y22, 
                                          hacia_entrada = True,
                                          k_gap_width = 0.5)
dib.display(d)
```

Obtuve los mismos resultados que en los cálculos

No puede simularse de una manera visual en LTSpice dado que es una red disipativa y tiene las singularidades en el eje sigma (excitación exponencial)

#### Simulaciones Simbólicas

```python
from pytc2.general import print_latex, print_subtitle, a_equal_b_latex_s
import sympy as sp
import pytc2.remociones as rmv
import pytc2.dibujar as dib
 
# Resolución simbólica
s = sp.symbols('s ', complex=True)

# función de excitación
Z11=(s**2+6*s+8)/(s**2+4*s+3)

# remociones (desde atras hacia adelante xq son parametros Y):    
print_subtitle("Remuevo R serie")
Z2, R1 = rmv.remover_valor_en_infinito(Z11)
print_latex(a_equal_b_latex_s("Z_2",Z2))
print_latex(a_equal_b_latex_s("R_1",R1))

print_subtitle("Remuevo RC serie")
Z4, Z3, R2, C2 = rmv.remover_polo_sigma(Z2, 1,isImpedance=True, isRC=True)
print_latex(a_equal_b_latex_s("Z_4",Z4))
print_latex(a_equal_b_latex_s("R_2",R2))
print_latex(a_equal_b_latex_s("C_2",C2))

print_subtitle("Remuevo R paralelo")
Z6, G3 = rmv.remover_valor_en_dc(1/Z4)
R3=1/G3
print_latex(a_equal_b_latex_s("Z_6",Z6))
print_latex(a_equal_b_latex_s("R_3",R3))

print_subtitle("Remuevo C paralelo")
_,C4 = rmv.remover_polo_infinito(Z6)
print_latex(a_equal_b_latex_s("C_4",C4))

# dibujo red:
d = dib.Drawing(unit=4)
d = dib.dibujar_puerto_entrada(d,
                        voltage_lbl = ('+', '$V_1$', '-'), 
                        current_lbl = '$I_1$')
d, zz_lbl = dib.dibujar_funcion_exc_abajo(d, 
                                          'Z_{11}',  
                                          Z11, 
                                          hacia_salida = True,
                                          k_gap_width = 0.5)
d = dib.dibujar_elemento_serie(d, dib.Resistor, R1)
d = dib.dibujar_tanque_RC_serie(d, R2, C2)
d = dib.dibujar_elemento_derivacion(d, dib.Resistor, R3)
d = dib.dibujar_espacio_derivacion(d)
d = dib.dibujar_elemento_derivacion(d, dib.Capacitor, C4)
d = dib.dibujar_puerto_salida(d,
                        voltage_lbl = ('+', '$V_2$', '-'), 
                        current_lbl = '$I2$')
dib.display(d)
```

Obtuve los mismos resultados que en los cálculos
