### Ejercicio 1: 
Hallar $\dfrac{V2}{V1}$

Lo resuelvo por superposición $V2 = V2'+V2''$

Defino: $k = \dfrac{r_3}{r_3+\frac{1}{cs}} = \dfrac{csr_3}{csr_3+1}$

$ V_2' = V_1 k (1+\frac{r_2}{r_1})$

$ V_2'' = V_1 (-\frac{r_2}{r_1})$

$ V_2 = V_1 [k(1+\frac{r_2}{r_1})-\frac{r_2}{r_1}] = v_1(k+k\frac{r_2}{r_1}-\frac{r_2}{r_1})$

$ \frac{V_2}{V_1} = k+k\frac{r_2}{r_1}-\frac{r_2}{r_1}$

Cambio de variable $k=\frac{k_n}{k_d}=\frac{scr_3}{1+scr_3}$

$\Large \frac{V_2}{V_1} = \bbox[yellow]{\frac{r_1}{r_1}}\frac{k_n}{k_d}+\frac{k_n}{k_d}\frac{r_2}{r_1}-\frac{r_2}{r_1}\bbox[yellow]{\frac{k_d}{k_d}}$

$\Large \frac{V_2}{V_1} = \frac{k_nr_1+k_nr_2-k_dr_2}{r_1k_d}$

$\Large \frac{V_2}{V_1} = \frac{scr_3r_1+scr_3r_2-(csr_3+1)r_2}{r_1(csr_3+1)} = \frac{scr_3r_1-r_2}{csr_3r_1+r_1)}\bbox[yellow]{\frac {\frac{1}{cr_3r_1}} {\frac{1}{cr_3r_1}}}$
